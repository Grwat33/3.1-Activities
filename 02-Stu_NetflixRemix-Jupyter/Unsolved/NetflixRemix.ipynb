```python
# Modules
import os
import csv
```

```python
user_movie = input("What movie are you looking for? ")
```

```python
file_path = os.path.join('Resources', 'netflix_ratings.csv')
```

```python
with open(file_path, 'r') as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    
    for movie in reader:
        title = movie[0]
        if user_movie == title:
            print(movie)
```
